```python
# Install dependencies first (run in Colab cell)
# pip install google-play-scraper pandas


from google_play_scraper import Sort, reviews
import pandas as pd
from google.colab import files

APP_ID = 'com.zeptoconsumerapp'  # Zepto package name

all_reviews = []
next_token = None
MAX_REVIEWS = 350000  # Limit to 3.5 lakh

while len(all_reviews) < MAX_REVIEWS:
    result, next_token = reviews(
        APP_ID,
        lang='en',            # English reviews
        country='in',         # India
        sort=Sort.NEWEST,     # Sort by newest
        count=25000,            # Fetch 25000 at a time
        continuation_token=next_token
    )

    all_reviews.extend(result)
    print(f"Fetched {len(result)} reviews, total so far: {len(all_reviews)}")

    if not next_token:  # No more reviews available
        break

# Limit in case it fetched a bit over
all_reviews = all_reviews[:MAX_REVIEWS]

# Save to CSV
csv_filename = "zepto_reviews.csv"
df = pd.DataFrame(all_reviews)
df.to_csv(csv_filename, index=False, encoding="utf-8")
print(f"Saved {len(df)} reviews to {csv_filename}")

# Download the CSV in Colab
files.download(csv_filename)
```

```python
# Install dependencies first (run in Colab cell)
# pip install google-play-scraper pandas

from google_play_scraper import Sort, reviews
import pandas as pd
from google.colab import files

APP_ID = 'com.grofers.customerapp'  # Blinkit package name

all_reviews = []
next_token = None
MAX_REVIEWS = 350000  # Limit to 3.5 lakh

while len(all_reviews) < MAX_REVIEWS:
    result, next_token = reviews(
        APP_ID,
        lang='en',            # English reviews
        country='in',         # India
        sort=Sort.NEWEST,     # Sort by newest
        count=25000,            # Fetch 25000 at a time
        continuation_token=next_token
    )

    all_reviews.extend(result)
    print(f"Fetched {len(result)} reviews, total so far: {len(all_reviews)}")

    if not next_token:  # No more reviews available
        break

# Limit in case it fetched a bit over
all_reviews = all_reviews[:MAX_REVIEWS]

# Save to CSV
csv_filename = "blinkit_reviews.csv"
df = pd.DataFrame(all_reviews)
df.to_csv(csv_filename, index=False, encoding="utf-8")
print(f"Saved {len(df)} reviews to {csv_filename}")

# Download the CSV in Colab
files.download(csv_filename)
```
